# Lambda Funtion or Anonymous
Function without name is called Lambda function.
It start with any variable assignment.

```python
#lambda expresion or lambda function
f = lambda a : a * a # hear a is an argument & operation in the argument is a * a
result = f(5)
result

# hear anonymous function is called lambda 
# remember lambda always you need to assgin as function cuz function are object in python
```

```python
f = lambda a, b : a + b
f1 = lambda a, b: a-b

result = f(1,4)
result1 = f1(4,1)

print(result)
print(result1)
```

# Filter()

```python
nums = [3,2,6,8,4,6,2,9]

evens = list(filter(is_even,nums))
```

```python
def is_even(n):
    return n%2==0

nums =[3,2,6,8,4,6,2,9]
evens = list(filter(is_even,nums))
print(evens)
```

```python
def is_odd(n):
    return n%2!=0

nums =[3,2,6,8,4,6,2,9]
evens = list(filter(is_odd,nums))
print(evens)
```

```python
nums = [3,2,6,8,4,6,2,9]
                                           # reduce the code
evens= list(filter(lambda n: n%2==0,nums))
print(evens)
```

```python
nums = [3,2,6,8,4,6,2,9]
                                           
odd= list(filter(lambda n: n%2!=0,nums))
print(odd)
```

```python
nums = [3,2,6,8,4,6,2,9]
                                           
evens= list(filter(lambda n: n%2==0,nums))
odd= list(filter(lambda n: n%2!=0,nums))
print(evens)
print(odd)
```

# Map
it double the even number i.e 2 become 4||4 become 6||6 become 8

```python
def update(n):
    return n*2

nums = [3,2,6,8,4,6,2,9]

evens = list(filter(is_even,nums))
double = list(map(update,evens))

print(evens)
print(double)
```
